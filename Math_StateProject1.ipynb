```python

# Grouped statistics

Product_category = df.groupby('Product Category')[['Sales', 'Profit']].describe()

Product_category
```

```python
category_subcategory = df.groupby(['Product Category', 'Product Sub-Category'])['Sales'].sum()
category_subcategory
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
numeric_cols.corr()
```

```python
Total_profit_by_category = df.groupby('Product Category')['Profit'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 4))
Total_profit_by_category.plot(kind='bar', color='blue',width=0.35)
plt.title('Total Profit of Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend()
plt.show
```

```python
df[['Sales', 'Profit']].corr().loc['Sales', 'Profit']
```

```python
import seaborn as sns
plt.figure(figsize=(18,7))
sns.scatterplot(data=df, x='Sales', y='Profit', alpha=0.8,color='red')
plt.title('Scatter Plot of Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.legend()
plt.show
```

```python
sales_by_ship_mode = df.groupby('Ship Mode')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sales_by_ship_mode.plot(kind='bar', color='purple',width=0.46)
plt.title('Total Sales by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Total Sales')
plt.xticks(rotation=30)
plt.grid(axis='y')
plt.legend()
plt.show
```
